```python
def ck(a,b):
    c=0
    for i in range(len(b)):
        if b[i] in a:
            c=c+1
            a[a.index(b[i])]=-1
        else:
            c=0
            break
    if(c==len(b)):
        return(1)
    else:
        return(0)
```
