```python
def check(A):
    if len(set(A)) == 1:
        return True
    return False

def min_steps(N,A,level):
    # If all are equal return the level
    if N == 1 or check(A):
        return level
    
    # Initialize min variable
    mn = float('+inf')
    # Try merging all one by one and recur
    for i in range(N-1):
        temp = A[:]
        temp[i]+=temp[i+1]
        temp.pop(i+1)
        mn = min(mn, min_steps(N-1,temp, level+1))
    print(temp)
    return mn
```

```python
def findSubstrings(s, k): 
    ans = 0
    n = len(s) 
    for i in range(n): 
        cnt = [0] * 26
        for j in range(i, n): 
            cnt[ord(s[j]) - ord('a')] += 1
            if (cnt[ord(s[j]) - ord('a')] <= k): 
                ans += 1
            else: 
                break
    return ans 
 
t=int(input())
while t:
    t=t-1
    S,k =input().split()
    print(findSubstrings(S, int(k)))
```

```python
def lcs(X, Y, m, n):
    if m == 0 or n == 0: 
        return 0; 
    elif X[m-1] == Y[n-1]: 
        return 1 + lcs(X, Y, m-1, n-1); 
    else: 
        return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n)); 
```

```python
def isPrime(n) :
    if (n <= 1) :
        return False
    if (n <= 3) :
        return True
    if (n % 2 == 0 or n % 3 == 0) :
        return False
    i = 5
    while(i * i <= n) :
        if (n % i == 0 or n % (i + 2) == 0) :
            return False
        i = i + 6
    return True
```

```python
def db(num,base):
    base_num = ""
    while num>0:
        dig = int(num%base)
        if dig<10:
            base_num += str(dig)
        else:
            base_num += chr(ord('A')+dig-10) 
        num //= base

    base_num = base_num[::-1]
    return base_num
```

```python
def xprint(size):
    i,j = 0,size - 1

    while j >= 0 and i < size:

        initial_spaces = ' '*min(i,j)
        middle_spaces = ' '*(abs(i - j) - 1)
        final_spaces = ' '*(size - 1 - max(i,j))

        if j == i:
            print (initial_spaces + '*' + final_spaces+initial_spaces + '*' + final_spaces)
        else:
            print (initial_spaces + '*' + middle_spaces + '*' + final_spaces+initial_spaces + '*' + middle_spaces + '*' + final_spaces)

        i += 1
        j -= 1

xprint(9)
```
