```python
article = cnn_paper.articles[6]
```

```python
article.download()
article.parse()
article.nlp()
```

```python
dt=article.publish_date
dt=dt.strftime('%Y-%m-%d::%H-%M')
```

```python
aut=article.authors
```

```python
art=article.text
```

```python
article.nlp()
```

```python
ark=article.keywords
ark=list(map(lambda x:x+'-',ark))
ark
```

```python
ars=article.summary
```

```python
import xlwt 
from xlwt import Workbook 
```

```python
wb = Workbook() 
```

```python
sheet1 = wb.add_sheet('Sheet 1') 
```

```python
sheet1.write(0, 1, 'Date and Time Posted') 
sheet1.write(0, 2, 'Author') 
sheet1.write(0, 3, 'Keywords') 
sheet1.write(0, 4, 'Article content') 
sheet1.write(0, 5, 'Article summary')
```

```python
sheet1.write(1, 1, dt) 
sheet1.write(1, 2, aut) 
sheet1.write(1, 3, ark) 
sheet1.write(1, 4, art) 
sheet1.write(1, 5, ars) 
```

```python
wb.save('news_ex.xls') 
```
